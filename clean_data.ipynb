```python
import pandas as pd
```

```python
x=pd.read_csv("loan_dirty.csv")
```

```python
x
```

```python
#The number of missing values in each column
x.isnull().sum()
```

```python
#only annual income is numerical data.. we fill it up with mean
x_fill=x.fillna(x.mean())
```

```python
#Now number of missing values in annual_inc is zero
x_fill.isnull().sum()
```

```python
#filling up columns with missing categorical data values
#ffill is forward fill that fills the nan value with previous valid element
x_fill=x_fill.fillna(method="ffill")
```

```python
#Now no nan values in dataset
#All null values have been suitably filled
print(x_fill.isnull().sum())
```

```python
#Saving the cleaned dataset to a new csv file called loan_clean.csv
x_fill.to_csv("loan_clean.csv")
```
